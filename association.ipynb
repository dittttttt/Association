```python
# Mengimpor library yang diperlukan 
import numpy as np
import pandas as pd 
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
#Meload dan mengeksport data
#Menginput data
data = pd.read_csv("Customers.csv")
data.head()
```

```python
# mengeksplore kolom data
data.columns
```

```python
# mengeksplore macam macam profesi
data.Profession.unique()
```

```python
#Memisahkan data sesuai dengan profesi
#profesi tersebut adalah artist
Artist = (data[data['Profession'] == "Artist"]
                 .groupby(['Age', 'Spending_Score'])['Profession']
                 .count().unstack().reset_index().fillna(0)
                 .set_index('Age'))

#profesi tersebut adalah Marketing
Marketing = (data[data['Profession'] == "Marketing"]
                 .groupby(['Age', 'Spending_Score'])['Profession']
                 .count().unstack().reset_index().fillna(0)
                 .set_index('Age'))

#profesi tersebut adalah Lawyer
Lawyer = (data[data['Profession'] == "Lawyer"]
                 .groupby(['Age', 'Spending_Score'])['Profession']
                 .count().unstack().reset_index().fillna(0)
                 .set_index('Age'))

#profesi tersebut adalah Doctor
Doctor = (data[data['Profession'] == "Doctor"]
                 .groupby(['Age', 'Spending_Score'])['Profession']
                 .count().unstack().reset_index().fillna(0)
                 .set_index('Age'))
Doctor
```

```python
# pengkodean data dengan "hot encoding"
# mendefinisikan fungsi "hot encoding" untuk membuat data sesuai untuk library yang bersangkutan 

def hot_encode(x):
    if(x<= 0):
        return 0
    if(x>= 1):
        return 1
```

```python
# mengkode dataset
prof_encoded = Artist.applymap(hot_encode)
Artist = prof_encoded

prof_encoded = Marketing.applymap(hot_encode)
Marketing = prof_encoded

prof_encoded = Lawyer.applymap(hot_encode)
Lawyer = prof_encoded

prof_encoded = Doctor.applymap(hot_encode)
Doctor = prof_encoded

Artist
```

```python
# memebangun model dan menganalisis
# 1) Artist 
# membangun model 
frq_items = apriori(Artist, min_support= 0.005, use_colnames= True)

#Mengumpulkan aturan yang disimpulkan dalam dataframe
rules = association_rules(frq_items, metric="lift", min_threshold= 1)
rules = rules.sort_values(['confidence', 'lift'], ascending = [False, False])
print(rules.head())
```

```python
# 2) Marketing 
# membangun model 
frq_items = apriori(Marketing, min_support= 0.005, use_colnames= True)

#Mengumpulkan aturan yang disimpulkan dalam dataframe
rules = association_rules(frq_items, metric="lift", min_threshold= 1)
rules = rules.sort_values(['confidence', 'lift'], ascending = [False, False])
rules.head()
```

```python
# 3) Lawyer 
# membangun model 
frq_items = apriori(Lawyer, min_support= 0.005, use_colnames= True)

#Mengumpulkan aturan yang disimpulkan dalam dataframe
rules = association_rules(frq_items, metric="lift", min_threshold= 1)
rules = rules.sort_values(['confidence', 'lift'], ascending = [False, False])
rules.head(10)
```

```python
# 4) Doctor
# membangun model 
frq_items = apriori(Doctor, min_support= 0.005, use_colnames= True)

#Mengumpulkan aturan yang disimpulkan dalam dataframe
rules = association_rules(frq_items, metric="lift", min_threshold= 1)
rules = rules.sort_values(['confidence', 'lift'], ascending = [False, False])
rules.head(3)
```
